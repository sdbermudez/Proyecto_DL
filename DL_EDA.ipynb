```python
import yfinance as yf
import pandas as pd

def download_price_data(
    tickers,
    start="2015-01-01",
    end=None
):
    """
    Descarga datos históricos de precios ajustados
    y los devuelve en formato largo (EDA-friendly).
    """
    data = yf.download(
        tickers,
        start=start,
        end=end,
        auto_adjust=True,
        progress=False
    )

    if data.empty:
        raise ValueError("No se pudieron descargar los datos.")

    # Pasar de formato ancho a largo
    data = (
        data
        .stack(level=1)
        .reset_index()
        .rename(columns={
            "level_1": "Ticker",
            "Date": "date",
            "Open": "open",
            "High": "high",
            "Low": "low",
            "Close": "close",
            "Volume": "volume"
        })
        .sort_values(["Ticker", "date"])
    )

    return data
```

```python
TICKERS = ["EC", "TGLS", "CIB", "CMTOY"]

df = download_price_data(TICKERS)

df.tail(n=10)
```
